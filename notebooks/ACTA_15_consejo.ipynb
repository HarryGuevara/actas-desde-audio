```python
import whisper
import os
from google.colab import files

# This part is assumed to have been run previously to upload the file
# uploaded = files.upload()
# Once uploaded, the 'uploaded' dictionary contains the file(s) name(s) as keys.

# Get the name of the uploaded file(s). Assuming only one file is uploaded.
# If multiple files are uploaded, you might need to choose the correct one.
if uploaded:
    # Get the first filename from the uploaded dictionary keys
    AUDIO_PATH = list(uploaded.keys())[0]
else:
    # Handle the case where no file was uploaded
    raise FileNotFoundError("No file was uploaded using files.upload()")


# Verificar si el archivo existe
if not os.path.exists(AUDIO_PATH):
    # This check should now pass if the file was successfully uploaded
    raise FileNotFoundError(f"No se encontró el archivo: {AUDIO_PATH}")

# Cargar el modelo Whisper
print("Cargando modelo de Whisper...")
model = whisper.load_model("base")  # Puedes usar "small", "medium" o "large" si quieres más precisión

# Transcribir el audio
print(f"Transcribiendo: {AUDIO_PATH}")
result = model.transcribe(AUDIO_PATH, language="es")

# Mostrar transcripción
print("\n--- Transcripción del audio ---\n")
print(result["text"])

# Guardar transcripción en un archivo de texto
with open("transcripcion.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

print("\n✅ Transcripción completada y guardada en 'transcripcion.txt'")
```
